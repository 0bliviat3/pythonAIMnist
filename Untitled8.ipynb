```python
from tensorflow.keras.datasets import mnist
import numpy as np
(x_train,y_train),(x_test,y_test) = mnist.load_data()
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
import matplotlib.pyplot as plt
import random
for i in range(1,4,1):
    for j in range(1,4,1):
        plt.subplot(i,4,j)
        plt.imshow(x_train[random.randint(0,60000)],cmap="gray")
    plt.show()
```

```python
# 이미 train set과 test set으로 데이터를 받았지만
# 그냥 train set에서 1000개의 데이터를 7:3으로 나누어 학습, 테스트를 진행한다
x_test = x_train[700:1000]
y_test = y_train[700:1000]
x_train = x_train[:700]
y_train = y_train[:700]
```

```python
# 데이터 전처리
x_train, x_test = x_train/255.0, x_test/255.0
```

```python
import tensorflow as tf
# 모델 구성
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 모델 훈련
history = model.fit(x_train, y_train,epochs=700)

# 정확도 평가
test_loss, test_acc = model.evaluate(x_test, y_test)
print('테스트 정확도:', test_acc)
```

```python
loss = history.history["loss"]
acc = history.history["accuracy"]
plt.plot(range(len(loss)),loss,label = "Train Loss")
plt.plot(range(len(acc)),acc,label = "Train Accuracy")
plt.grid()
plt.legend()
plt.show()
```
